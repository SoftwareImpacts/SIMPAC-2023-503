# Data Preparation

```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv("oanda-SPX500_USD-2005-1.csv", index_col='time', parse_dates=True)
df1=pd.read_csv("oanda-NAS100_USD-2005-1.csv", index_col='time', parse_dates=True)
df2=pd.read_csv("oanda-FR40_EUR-2005-1.csv", index_col='time', parse_dates=True)
df3=pd.read_csv("oanda-AU200_AUD-2005-1.csv", index_col='time', parse_dates=True)
```

```python
df_list=[df,df1,df2,df3]
```

```python
df_list[0].head()
```

DataFrame should have the same structure as the DataFrame from Yahoo Finance:

Index: 'Date'

Columns: Close, High, Low, Open (the order is not important, yet they should have capital letters)

```python
for i in range(len(df_list)):
# renaming index
  df_list[i].index.names = ['Date']

# renaming columns
  df_list[i].rename(columns={"close":"Close", "high":"High", "low":"Low","open":"Open"}, inplace=True)

# removing unnecessary columns
  df_list[i].drop(columns=["volume"], inplace=True)
```

```python
df_list[3].head()
```

```python
df_csv=df_list[1].groupby(pd.Grouper(freq='5Min')).agg({
                                            "Close": "last","High": "last","Low": "last","Open": "last",
                                        })
df_csv.fillna(method='ffill', inplace=True)
df_csv.to_csv("oanda-NAS100_USD-2005-1_5min.csv")
```
